```python
#Задание 7
#Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def hypothesis_test(df):
    # Разделим сотрудников на группы по зарплате
    median_salary = df['salary'].median()
    high_income = df[df['salary'] >= median_salary]
    low_income = df[df['salary'] < median_salary]

    # Проверим нормальность распределения
    print("Проверка нормальности распределения для группы с высокой зарплатой:")
    print(stats.shapiro(high_income['average_montly_hours']))
    print("\nПроверка нормальности распределения для группы с низкой зарплатой:")
    print(stats.shapiro(low_income['average_montly_hours']))


    # Визуализация распределения
    plt.figure(figsize=(12, 6))
    sns.kdeplot(data=high_income, x='average_montly_hours', label='Высокая зарплата', fill=True)
    sns.kdeplot(data=low_income, x='average_montly_hours', label='Низкая зарплата', fill=True)
    plt.legend()
    plt.title('Распределение рабочего времени по группам зарплат')
    plt.show()

    # Проверим равенство дисперсий
    print("\nПроверка равенства дисперсий:")
    print(stats.levene(high_income['average_montly_hours'],
                      low_income['average_montly_hours']))

    # Выполним t-тест
    t_stat, p_value = stats.ttest_ind(high_income['average_montly_hours'],
                                     low_income['average_montly_hours'],
                                     equal_var=True)

    print(f"\nT-статистика: {t_stat}")
    print(f"P-значение: {p_value}")

    print("p > 0.05, значит, статистически значимых различий в среднем рабочем времени между двумя группами не обнаружено.")

    # Вывод результатов
    if p_value < 0.05:
        print("\nГипотеза отвергается: есть статистически значимая разница в рабочем времени между группами.")
    else:
        print("\nГипотеза не отвергается: статистически значимой разницы в рабочем времени между группами нет.")


# Пример использования:
if __name__ == "__main__":
    df = pd.DataFrame({
        'salary': np.random.randint(2000, 10000, 100),
        'average_montly_hours': np.random.randint(100, 200, 100)
    })

    hypothesis_test(df)
```

```python
#Задание 8
#Рассчитать следующие показатели среди уволившихся и не уволившихся сотрудников (по отдельности):
#● Доля сотрудников с повышением за последние 5 лет
#● Средняя степень удовлетворенности
#● Среднее количество проектов

def calculate_metrics(group):
    # Доля сотрудников с повышением за последние 5 лет
    promoted = group['promotion_last_5years'] == 1
    promoted_count = group[promoted].shape[0]
    employees_count = group['promotion_last_5years'].shape[0]
    promotion_rate = 0 if promoted.any() and employees_count == 0 else group[promoted].shape[0] / employees_count

    # Средняя степень удовлетворенности
    avg_satisfaction = group['satisfaction_level'].mean()

    # Среднее количество проектов
    avg_projects = group['number_project'].mean()

    return {
        'Доля с повышением за последние 5 лет': promotion_rate,
        'Средняя степень удовлетворенности': avg_satisfaction,
        'Среднее количество проектов': avg_projects
    }

# Расчет для уволившихся
metrics_left = calculate_metrics(df_left)

# Расчет для неуволившихся
metrics_not_left = calculate_metrics(df_not_left)

print("Уволившиеся сотрудники:")
for key, value in metrics_left.items():
    print(f"{key}: {value:.2f}")

print("\nНеуволившиеся сотрудники:")
for key, value in metrics_not_left.items():
    print(f"{key}: {value:.2f}")
```
